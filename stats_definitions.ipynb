# GLOSSARY

**Offensive Stats**

* ufga - number of field goals ATTEMPTED when the shooter does not have a defender within his vicinity (3.5 ft)
**ufgm - number of field goals MADE when the shooter does not have a defender within his vicinity (3.5 ft)**
* pts_paint - points in the paints
* ufgp - uncontested field goal percent? 
* FGM - field goals made
* FGA - field goal attempts
* FG % - field goal percent
* 2pm - two points made
* 2pa - two point attempts
* 2p% - two point percent
* 3PM - three points made
* 3PA - three point attempts
* 3P% - three point percentage
* FTM - free throw made
* FTA - free throw attempts
* FT% - free throw percent
* OREB - offensive rebounds
* REB - rebounds
**AST - assists** 
* TOV - turn overs
* PF - personal fouls
* PFD - percent of team's personal fouls drawn
* +/- - point differential when a player or team is on the floor
* Uncontested Assists - Assists that result in uncotested shots? 
* MIN - minutes
**OFFRTG - Measures a team's points scored per 100 possesions**
**NETRTG - Measures a team's point differential per 100 possesions**
* AST% - Percentage of teammate field goals a player assisted on while he was on the floor
* AST/TO - Assist to Turnover Ratio, number of assits for a player or team compared to # of turnovers they commmited
* AST RAT - Assist Ratio, number of assists a player avgs per 100 posessions used
* OREB% - Offensive Rebound Percetage, percentage of available offensive rebounds a team obtains while on floor
**EFG % - Effective Field Goal Percentage, FG% adjusted for made 3pt fg being 1.5x more valuable than 2pt fg.**
**TS% - True shooting percentage, factors value of 3pt fg and ft in addition to 2pt fg**
* PACE - Number of possesions per 48 minutes for a team or player
**PIE - Player Impact Estimate, measures a player's overall statistical contribution against total statistics in games**
* OPP PTS 2ND CHANCE - number of points scored on offensive rebound
* OPP PTS FB - Opponent Fast Break Points, points scored while on a fast break
* OPP PTS PAINT - Opponent Points in the Paint

* BLKA - Opponent Blocks: number of times an opponent has registred a block on the shot of a player or team
* OPP PTS OFF TOV - number of points scored by an opposing player or team following a turnover

**Defensive Stats**
* points_off_tov - points off turnovers
* DREB - defensive rebounds
* STL - steals
* BLK - blocks
* DEFRTG -  Number of points allowed per 100 possesions by a team
