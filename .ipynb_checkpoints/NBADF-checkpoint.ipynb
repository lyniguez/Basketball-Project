```python
import pandas as pd
import numpy as np
```

```python
advTeamStats = pd.read_csv("Basketball_Stats/advanced_team_stats.csv")
finalSVUStats = pd.read_csv("Basketball_Stats/final_svu_teams.csv")
miscTeamStats = pd.read_csv("Basketball_Stats/misc_team_stats.csv")
```

```python
advTeamStatsDF = advTeamStats.drop_duplicates()
advTeamStatsDF.info()
```

```python
finalSVUStats.info()
```

```python
miscTeamStats.info()
```

```python
advmiscteam = advTeamStats.merge(miscTeamStats,on="game_id")
```

```python
advmiscsvuDF = advmiscteam.merge(finalSVUStats,on="game_id")
```

```python
teamsDF = advmiscsvuDF[['game_id','team','season','off_rating','def_rating','ufga','ufgm','pts_off_tov','pts_paint']]
teamsDF
```

```python
teamsDF2 = teamsDF.groupby(["team","game_id","season"]).mean()
```

```python
teamsDF3 = teamsDF2.reset_index()
teamsDF3
```

```python
teamsDF4 = teamsDF3[['team','season','off_rating','def_rating','ufga','ufgm','pts_off_tov','pts_paint']]
teamsDF4
```

df1 = teamsDF4.loc[teamsDF4["team"]=="Bulls"]

writer = pd.ExcelWriter('NBA.xlsx')
df1.to_excel(writer,'Sheet1')


```python
teamsDF5 = teamsDF4.groupby(["team","season"]).mean()
teamsDF6 = teamsDF5.reset_index()
teamsDF6["ufgp"] = teamsDF6["ufgm"] / teamsDF6["ufga"]
teamsDF6.sort_values("ufgp",ascending = False)
```

```python
df1 = teamsDF5
writer = pd.ExcelWriter('NBADF.xlsx')
df1.to_excel(writer,'Sheet1')
```
